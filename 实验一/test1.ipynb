```python
import pandas as pd
from pandas import DataFrame
import numpy as np

primitive_data=pd.read_csv("data.csv",encoding='gbk')
primitive_data#原始数据，实际测试已经没有空行
```

```python
primitive_data_1=primitive_data.dropna(how='any')
primitive_data_1#原始数据已经过滤空行，与primitive_data一致
```

```python
#过滤得到traffic不等于0且from_level=一般节点的数据
data_before_filter=primitive_data_1
data_after_filter_1=data_before_filter.loc[data_before_filter["traffic"]!=0]
data_after_filter_2=data_after_filter_1.loc[data_after_filter_1["from_level"]=='一般节点']
data_after_filter_2
```

```python
#加权抽样
data_before_sample=data_after_filter_2
columns=data_before_sample.columns
weight_sample=data_before_sample.copy()
weight_sample['weight']=0
for i in weight_sample.index:
    if weight_sample.at[i,'to_level']=='一般节点':
        weight=1
    else:
        weight=5
    weight_sample.at[i,'weight']=weight

weight_sample_finish=weight_sample.sample(n=50,weights='weight')

weight_sample_finish=weight_sample_finish[columns]
weight_sample_finish
```

```python
#随机抽样
random_sample=data_before_sample
random_sample_finish=random_sample.sample(n=50)
random_sample_finish=random_sample_finish[columns]
random_sample_finish
```

```python
#分层抽样
ybjd=data_before_sample.loc[data_before_sample['to_level']=='一般节点']
wlhx=data_before_sample.loc[data_before_sample['to_level']=='网络核心']
after_sample=pd.concat([ybjd.sample(17),wlhx.sample(33)])
after_sample
```

```python
#实现系统抽样
data_before_sample = data_after_filter_2
columns = data_before_sample.columns

def systematic_sampling(data, sample_size):
    total_rows = len(data)

    interval = int(np.ceil(total_rows / sample_size))
    sample_indices = np.arange(0, total_rows, interval)[:sample_size]
    return data.iloc[sample_indices].copy()

systematic_sample = systematic_sampling(data_before_sample, sample_size=50)
systematic_sample_finish = systematic_sample[columns]

systematic_sample_finish
```

```python
#以to_city作为"群"的划分依据,实现整群抽样）
import random
data_before_sample = data_after_filter_2
columns = data_before_sample.columns
clusters = data_before_sample['to_city'].unique()
print(f"所有群标识：{clusters}")
selected_clusters = random.sample(list(clusters), k=10)  # 抽取10个群
print(f"随机选中的群：{selected_clusters}")
cluster_sample = data_before_sample[
    data_before_sample['to_city'].isin(selected_clusters)
].copy()

cluster_sample_finish = cluster_sample[columns]

cluster_sample_finish
```
